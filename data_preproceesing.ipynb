```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('/content/ML.csv')
data.head()
```

```python
data.describe()
```

```python
data['Salary']=data['Salary'].fillna(data['Salary'].mean())
data['CGPA']=data['CGPA'].fillna(data['CGPA'].mean())
data.head()
```

```python
cgpa=data[['CGPA']]
salary=data['Salary']
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model.fit(cgpa,salary)
result=model.predict([[10]])
print("Predicted Salary:",result)
```
