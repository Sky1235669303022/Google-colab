```python
import tensorflow as tf  #Name: Sudhanshu Kukreti
import numpy as np       #Erp :0231BCA041
```

```python
gender= np.array([[6,60],[7,60], #for male
                 [5,50],[4,50]])  # for female
label=np.array([[1],[1],[0],[0]])
```

```python
model= tf.keras.Sequential([tf.keras.layers.Dense(1,activation='sigmoid',input_shape=(2,))])
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.fit(gender,label,epochs=100,verbose=1)
```

```python
result=model.predict(np.array([[4,55]]))
print((result > 0.5).astype(int))
```
